# i)treating the null values(fill with mean or median or mode)

```python
import pandas as pd

# Load the dataset
file_path = "/content/drive/MyDrive/online_shoppers_intention (2).csv"
df = pd.read_csv(file_path)

# Fill null values: mean for numerical, mode for categorical
df.fillna(df.mean(numeric_only=True), inplace=True)  # Mean for numerical columns
df.fillna(df.mode().iloc[0], inplace=True)  # Mode for categorical columns

# Check for remaining null values
print(df.isnull().sum())

```

ii)drop null values if they are less than 5%-10%


```python
import pandas as pd

# Load the dataset
file_path = "/content/drive/MyDrive/online_shoppers_intention (2).csv"
df = pd.read_csv(file_path)

# Define the threshold for dropping columns (5% to 10% missing values)
threshold = 0.10  # Change to 0.05 for 5%

# Calculate the percentage of missing values in each column
missing_percent = df.isnull().sum() / len(df)

# Drop columns with missing values less than the threshold
df = df.dropna(subset=missing_percent[missing_percent.between(0.05, threshold)].index)

# Fill remaining null values: mean for numerical, mode for categorical
df.fillna(df.mean(numeric_only=True), inplace=True)  # Mean for numerical columns
df.fillna(df.mode().iloc[0], inplace=True)  # Mode for categorical columns

# Check if any null values remain
print(df.isnull().sum())
```

#9)treating the ouliers
An outlier is a data point that is very different from the rest of the data. It is either too high or too low compared to most values.


```python
# Remove outliers using IQR
for col in df.select_dtypes(include=[np.number]).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Print dataset shape after removing outliers
print(df.shape)
```

## ii)remove outliers

```python
# Remove outliers using IQR
for col in df.select_dtypes(include=['number']).columns:
    Q1 = df[col].quantile(0.25)  # 25th percentile
    Q3 = df[col].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile Range

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]  # Keep only valid values

# Print the new shape of the dataset
print(f"Dataset shape after outlier removal: {df.shape}")
```

```python
#K-mean
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/online_shoppers_intention (2).csv")

# Handling missing values
df.dropna(inplace=True)

# Encoding categorical variables if any
categorical_cols = df.select_dtypes(include=['object']).columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Feature Scaling
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Finding the optimal number of clusters using the Elbow Method
inertia = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()

# Training K-Means with optimal clusters (assume k=3 for now, adjust as needed)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(df_scaled)

# Visualizing the Clusters using PCA for dimensionality reduction
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
df_pca['Cluster'] = df['Cluster']

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='Cluster', palette='viridis')
plt.title('K-Means Clustering Visualization')
plt.show()

# Display Cluster Centers
print("Cluster Centers (Scaled):")
print(kmeans.cluster_centers_)
```

3)Hyperparameter Tuning (Optimizing Number of Clusters, etc.)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/online_shoppers_intention (2).csv")

# Handling missing values
df.dropna(inplace=True)

# Encoding categorical variables if any
categorical_cols = df.select_dtypes(include=['object']).columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Feature Scaling
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Hyperparameter Tuning for K-Means (Finding Optimal Number of Clusters using Elbow Method and Silhouette Score)
inertia = []
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(df_scaled)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(df_scaled, labels))

# Plot Elbow Method
plt.figure(figsize=(8, 4))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()

# Plot Silhouette Score
plt.figure(figsize=(8, 4))
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')
plt.show()

# Choosing optimal K
optimal_k = k_range[np.argmax(silhouette_scores)]
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(df_scaled)

# Hyperparameter Tuning for DBSCAN (Finding Optimal eps using k-distance plot)
from sklearn.neighbors import NearestNeighbors

neigh = NearestNeighbors(n_neighbors=5)
neigh.fit(df_scaled)
distances, indices = neigh.kneighbors(df_scaled)
distances = np.sort(distances[:, -1])

plt.figure(figsize=(8, 4))
plt.plot(distances)
plt.xlabel('Data Points Sorted by Distance')
plt.ylabel('5th Nearest Neighbor Distance')
plt.title('Elbow Method for DBSCAN eps')
plt.show()

# DBSCAN Clustering with optimized parameters
dbscan = DBSCAN(eps=1.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(df_scaled)

# Hierarchical Clustering
linkage_matrix = linkage(df_scaled, method='ward')
hierarchical_labels = fcluster(linkage_matrix, 3, criterion='maxclust')

# Evaluating Clusters
def evaluate_clusters(data, labels, method_name):
    if len(set(labels)) > 1:
        silhouette_avg = silhouette_score(data, labels)
        davies_bouldin = davies_bouldin_score(data, labels)
        calinski_harabasz = calinski_harabasz_score(data, labels)
        print(f"{method_name} Clustering Evaluation:")
        print(f"  Silhouette Score: {silhouette_avg:.4f}")
        print(f"  Davies-Bouldin Index: {davies_bouldin:.4f}")
        print(f"  Calinski-Harabasz Index: {calinski_harabasz:.4f}\n")
    else:
        print(f"{method_name} Clustering: Only one cluster detected, cannot compute metrics.\n")

# Print evaluation for each method
evaluate_clusters(df_scaled, kmeans_labels, "K-Means")
evaluate_clusters(df_scaled, dbscan_labels, "DBSCAN")
evaluate_clusters(df_scaled, hierarchical_labels, "Hierarchical")

# Visualizing Clusters using PCA
def plot_clusters(labels, title):
    pca = PCA(n_components=2)
    df_pca = pca.fit_transform(df_scaled)
    df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
    df_pca['Cluster'] = labels

    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='Cluster', palette='viridis')
    plt.title(title)
    plt.show()

# Plot for each clustering method
plot_clusters(kmeans_labels, "K-Means Clustering Visualization")
plot_clusters(dbscan_labels, "DBSCAN Clustering Visualization")
plot_clusters(hierarchical_labels, "Hierarchical Clustering Visualization")
```
